# Estimating cloud cover
I am using the cloud cover each night to quantify observing luck. McDonald Observatory has a cloud sensor, but it has a lot of missing data. The purpose of this notebook is to try to fill in the missing data using humidity and temperature.

```python
import numpy as np
import pandas as pd
import datetime
from astropy import time
%matplotlib inline
import pylab
import seaborn as sns
sns.set_context('poster')
pd.set_option('use_inf_as_null', True)
```

Read in the cloud data. We need to take care for null values. We will also parse the 2nd and 3rd column to get a date column in a nice format. 

```python
# Read in all data
cloud_data = pd.read_csv('data/clouds.dat', na_values='+-.----E+--', parse_dates={"Date": [1,2]})
basic_data = pd.read_csv('data/basic_weather.dat', na_values='+-.----E+--', parse_dates={"Date": [0]})
basic_data.index = basic_data.Date
cloud_data.index = cloud_data.Date

# Store the non-null data in a new dataframe
good = cloud_data.dropna(subset=['cloud-RH', 'cloud-Sky-ambient'], how='any')
```

We need to fill in some missing values in the basic weather data. These are generally less of an issue than the missing values in the cloud sensor data, so we don't need to try quite as hard. There are a few rows where only the dew point is missing. Since that [is calculable](http://iridl.ldeo.columbia.edu/dochelp/QA/Basic/dewpoint.html) from everything else, we will just re-calculate. If the other things are nan's as well, the result will be a nan which is fine. We will address that in the next step by simply interpolating along columns.

```python
# Calculate dew point
T = (basic_data['base-TEMP_AVG'] - 32.0)*5.0/9.0 + 273
Es = np.exp(5423*(1.0/273. - 1.0/T))
Td = (1./273. - (1./5423.)*np.log(basic_data['base-R.H._AVG']*Es/100.))**-1
basic_data['DEWPT_calculated'] = (Td - 273.)*(9./5.) + 32
```

```python
# Interpolate all values that are still null in time
basic_data.interpolate(method='time', inplace=True)
```

```python
# Save the interpolated basic data
basic_data.to_csv('data/basic_weather_fixed.csv', index=False)
```

Now, the cloud sensor works by taking the difference of the ambient temperature (in the cloud-Ambient column) from the sky temperature (in the cloud-Sky_cond) column. The sky temperature is measured from the 8-15 $\mu m$ flux. The closer those two values are to each other, the more cloud cover. We will use three categories here, with thresholds that come from the McDonald Observatory [weather site](http://weather.as.utexas.edu/weather_graph/Cloudy.html).

  - $\Delta T < -54$: Clear
  - $-54 < \Delta T < -18$: Partly cloudy
  - $\Delta T > -18$: Very cloudy

```python
# Map the sky-ambient column to cloud cover levels
def get_cloudiness(Tdiff):
    if Tdiff < -54:
        return 0
    elif -18 > Tdiff >= -54:
        return 1
    else:
        return 2
good['cloudiness'] = good['cloud-Sky-ambient'].map(get_cloudiness)
cloud_data['cloudiness'] = cloud_data['cloud-Sky-ambient'].map(get_cloudiness)
```

Let's do a little bit of exploratory data analysis. We will bin the humidity and temperature values, and color-code by an average cloud cover using the new 'cloudiness' column.

```python
out = pylab.hexbin(good['cloud-RH'].values, good['cloud-Ambient'].values, C=good['cloudiness'],
                   bins=None, extent=(0, 100, -50, 130))
cb = pylab.colorbar()
pylab.xlabel('Relative Humidity')
pylab.ylabel('Ambient temperature (F)')
cb.set_label('Average Cloudiness')
pylab.savefig('Figures/Cloud_Data.png')
```

It is pretty clear from this plot that the heaviest cloud cover comes when humidity is high, or the ambient temperature is high. Since humidity is a function of temperature, it would probably make sense to use precipitable water vapor instead. We will need pressure for that. Nonetheless, things looks promising: there is definitely some relationship between temperature, water vapor, and cloud cover that a machine learning algorithm should be able to tease out.

## Predicting cloudiness
Now, we move on to predicting the cloudiness of a given night from the basic data. Based on the exploratory analysis we did above, the variables we will use are:

  - Temperature
  - Dew point
  - Pressure
  - Change in relative humidity

and the thing we want is the cloudiness variable, which can take one of three values:

  - 1: Clear
  - 2: Partly cloudy
  - 3: Very cloudy

We will test a few different models, and see what gives the best results. We will measure the results by training the model on 90% of the data, and cross-validating on the final 10%. Here are the things we will try:

### [Logistic regression](http://en.wikipedia.org/wiki/Logistic_regression#Logistic_function.2C_odds.2C_odds_ratio.2C_and_logit): 

Logistic regression is easiest to understand when there are only two classes instead of the three we have here. Then, it is similar to polynomial regression in that you are fitting an equation of the form

$ t = c_0 + c_1x_1 + c_2x_2 + ...$

 to the data (x), but then taking the extra step of turning t into a probability of being in 'class 1' with the logistic function:

$ f(t) = \frac{1}{1+e^{-t}}$

Thus, you fit the parameters $c_0, c_1, ...$, and evaluate the parameters by comparing the value of f(t) to what class the data point is in. The optimized parameters then have the properties:

  - points in class 1: f(t) > 0.5 (or some threshold you can play with if you want)
  - points in class 0: f(t) < 0.5
  
To use the logistic function to predict the class, you just plug your data points in to get t, and then use the logistic function to get the probability of being in class 1.

How do we use this for the problem here, where there are three classes? The simplest way is the 'one vs. all' classifier. Basically, you fit three sets of parameters $c_0, c_1, ...$ and each set corresponds to one of the classes. In our setup, you would do normal logistic regression for:

  1. clear vs. partly cloudy *or* very cloudy
  2. partly cloudy vs. clear *or* very cloudy
  3. very cloudy vs. clear *or* partly cloudy.
  
Then when you want to predict which class a new data point has (how cloudy is it), you find the probabilities of each class and take the one with the highest probability as the predicted class.


### [Support Vector Machine](https://www.youtube.com/watch?v=1NxnPkZM9bc)

A support vector machine (SVM) is another way to do classification. Again, it is easiest to understand in the two-class case and generalizes in exactly the same way as logistic regression. A SVM tries to directly fit a line that separates the two classes. The best line is that which has the most distance between points on the two classes. Using some mathematical tricks, the data points can be [transformed](https://www.youtube.com/watch?v=3liCbRZPrZA) such that a plane can separate groups that cannot be separated by a simple line. 

Why is it called a support vector machine? Well, it comes from the fact that only some of the points really play a role in setting where the dividing line goes. Since a SVM maximizes the margin between the dividing line and the nearest points in each class, the points that are far from the dividing line don't really matter. Those points that are close to the boundary are called the support vectors.

One major problem with support vector machines is that they are slow for large datasets. The implementation in scikit-learn has a computational complexity of O(N$^2$) to O(N$^3$), depending on the dataset. That means that the computational time increases very quickly with the number of points. My dataset has roughly 100,000 points in it, and the full cross-validation takes several hours. The same thing takes a couple minutes for the logistic regression.

### [Decision Tree](http://scikit-learn.org/dev/modules/tree.html)

A decision tree is probably the easiest to conceptualize. It is basically just a flowchart that leads to the classification labels (in our case clear, partly cloudy, and very cloudy). Or, in programmer-speak, it is a long chain of if-then-else decision rules. An example assessing how likely a person was to have survived the Titanic is shown below.
<img src="Images/CART_tree_titanic_survivors.png" >

Fitting a decision tree from data is done by splitting each input variable (temperature, pressure, etc) at some value. The number of branches and numerical value of the splits are adjusted until the probability of a class no longer changes by much at each leaf.

The chief advantage of a decision tree is that it is very easy to interpret. For this reason, they are popular machine learning tools to provide actionable results.


### [Random Forest Classifier](http://scikit-learn.org/dev/modules/ensemble.html#forests-of-randomized-trees)

A random forest is an improvement upon a decision tree. Basically, a bunch of decision trees are all trained at once with random subsamples of the data. To make a prediction, the data is sent through each tree and the total probability is calculated from the ensemble. As you add more trees to the "forest," the result gets more accurate but is also more prone to over-fitting. To find the best number of trees, we will cross-validate within the training sample to find the number that generalizes the best.

### [K-nearest neighbors](http://scikit-learn.org/dev/modules/neighbors.html#nearest-neighbors-classification)

The K-nearest neighbors algorithm is really easy to understand as well: to classify a new data point, you just take several of the points with known labels closest to it, and do a majority vote. If most of the points near the near data point are cloudy, for instance, then the new data point will be classified as cloudy. Unlike the other algorithms, K nearest neighbors does not build up any kind of global model that gets applied to the data. Instead, it just uses the training data itself. As with the random forest classifier, we will use cross-validation to choose the 'best' number of neighbors.

```python
data = pd.read_csv('data/final_cloud_data.csv')
```

```python
import sklearn
from sklearn import cross_validation
from sklearn import linear_model
from sklearn import preprocessing
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
import pydot
from sklearn.externals.six import StringIO
```

```python
# Get all the non-null data
X_df = data.loc[(data['cloud-Daylight'] < 1e6)][['base-TEMP_AVG', 'base-PRESS_AVG', 'DEWPT_calculated', 'RH_Change', 'T_Change', 'Dust']].dropna(how='any')
X_df = data.loc[(data['cloud-Daylight'] < 1e6)][['base-TEMP_AVG', 'base-PRESS_AVG', 'DEWPT_calculated', 'Dust']].dropna(how='any')
y_df = data.ix[X_df.index]['cloudiness']

# Normalize the data (make it so all the variables have roughly the same dynamic range)
X = X_df.values
scaler = preprocessing.StandardScaler().fit(X)
#X_norm = preprocessing.normalize(X)
X_norm = scaler.transform(X)
#print X_norm
#print X_norm2

# Split into a training and validation set
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X_norm, y_df.values, test_size=0.1)
#X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y_df.values, test_size=0.1)
```

```python
# Run the logistic regression
logistic = linear_model.LogisticRegressionCV(Cs=25)

logistic.fit(X_train, y_train)
log_score = logistic.score(X_test, y_test)
print 'Logistic regression score on the validation data: {:.3f}'.format(log_score)
```

```python
# Run a decision tree classifier
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
dtree_score = clf.score(X_test, y_test)
print 'Decision tree score on the validation data: {:.3f}'.format(dtree_score)
```

```python
#Plot the tree
export_graphviz(clf, out_file='tree.dot', feature_names=['T', 'P', 'Dew Point', r'$\Delta$ RH', r'$\Delta$ T', 'Dust'])
```

```python
# Get information about the tree and plot the importance of each feature.
print('Number of levels in full tree: {}'.format(clf.tree_.max_depth))
clf.feature_importances_
#tree_df = pd.DataFrame(data={'Features': ['T', 'P', 'Dew Point', 'RH change', 'T change', 'Dust count'],
#                             'Importance': clf.feature_importances_})
tree_df = pd.DataFrame(data={'Features': ['T', 'P', 'Dew Point', 'Dust count'],
                             'Importance': clf.feature_importances_})
x_order = tree_df.sort('Importance', ascending=False)['Features']
print tree_df
print x_order.values
p = sns.factorplot('Features', y='Importance', data=tree_df, kind='bar', size=10, x_order=x_order.values)
#p.savefig('Figures/Decision_Tree_Importance.png')
```

```python
# Cross-validate a random forest classifier with 3-1000 trees
N = np.logspace(0.5, 3, 25, dtype=np.int)
scores_arr = []
err_arr = []
for i, n in enumerate(N):
    clf = RandomForestClassifier(n_estimators=n, n_jobs=2)
    scores = cross_validation.cross_val_score(clf, X_train, y_train, cv=3)
    scores_arr.append(np.mean(scores))
    err_arr.append(np.std(scores))
    print('CV Score for N = {} trees = {}'.format(n, scores_arr[-1]))
    
```

```python
# Plot the cross-validation accuracy as a function of the number of trees
fig, ax= pylab.subplots()
ax.errorbar(N, scores_arr, fmt='o-', yerr=err_arr)
ax.set_xscale('log')
pylab.xlabel('Number of trees')
pylab.ylabel('Cross-validation accuracy')
pylab.title('Random Forest Classification')
pylab.savefig('Figures/RandomForest_CV.png')
```

```python
# Re-run the random forest with the best number of trees
#n = N[np.argmax(scores_arr)]
n = 100 # approximate location of the 'point of diminishing returns'
print('Number of trees in the random forest: {}'.format(n))
clf = RandomForestClassifier(n_estimators=n)
clf.fit(X_train, y_train)
rf_score = clf.score(X_test, y_test)
print 'Random forest score on the validation data: {:.3f}'.format(rf_score)
```

```python
# Cross-validate a K-nearest neighbors classifier with 3-1000 neighbors
N = np.logspace(0, 3, 25, dtype=np.int)
scores_arr = []
err_arr = []
for i, n in enumerate(N):
    clf = KNeighborsClassifier(n_neighbors=n, weights='distance')
    scores = cross_validation.cross_val_score(clf, X_train, y_train, cv=10)
    scores_arr.append(np.mean(scores))
    err_arr.append(np.std(scores))
    print('CV Score for N = {} neighbors = {}'.format(n, scores_arr[-1]))
```

```python
# Plot the cross-validation accuracy as a function of the number of neighbors
fig, ax= pylab.subplots()
ax.errorbar(N, scores_arr, fmt='o-', yerr=err_arr)
ax.set_xscale('log')
pylab.xlabel('Number of neighbors')
pylab.ylabel('Cross-validation accuracy')
pylab.title('K-Nearest Neighbors Classification')
pylab.savefig('Figures/KNN_CV.png')
```

What we see here is an example of 'bias,' or under-fitting. For more than about 50 neighbors, the majority vote actually becomes less useful. The low-N side of the curve is 'variance', or over-fitting. This is where there are too few neighbors, so it fits almost everything but does not generalize to new points very well.

```python
# Re-run the K-nearest neighbors classifier with the best number of neighbors
n = N[np.argmax(scores_arr)]
n=4
print('Number of neighbors in the KNN classifier: {}'.format(n))
clf = KNeighborsClassifier(n_neighbors=n, weights='distance')
clf.fit(X_train, y_train)
knn_score = clf.score(X_test, y_test)
print 'KNN score on the validation data: {:.3f}'.format(knn_score)
```

```python
def predict_cloudiness(clf, scale_fcn, df):
    """
    Given a trained classifier, a scaler function and a pandas dataframe, return a pandas series object with 
    the expected cloudiness.
    """
    X_df = df[['base-TEMP_AVG', 'base-PRESS_AVG', 'DEWPT_calculated', 'Dust']]
    X_norm = scale_fcn.transform(X_df.values)
    
    cloudiness_arr = clf.predict(X_norm)
    
    return pd.Series(data=cloudiness_arr, index=X_df.index)
```

```python
# Get the predicted cloudiness
data2 = data.copy()
cloudiness = predict_cloudiness(clf, scaler, data2.copy())

# Fill in the missing values
idx = data2.index[data2.cloudiness.isnull()]
data2.loc[idx, 'cloudiness'] = cloudiness.ix[idx]
```
